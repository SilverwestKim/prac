```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
grp1=pd.DataFrame(np.random.randn(20,7), columns=['X','Y','C','D','E','F','G'])
grp1.plot.scatter(x='X',y='Y')
```

```python
grp2=pd.Series(np.random.randn(1000))
grp2.plot()
```

```python
grp2 = pd.Series(np.random.randn(1000))
grp3 = grp2.cumsum()
grp3.plot()
```

```python
grp4=pd.Series(np.random.randn(1000))
grp4.plot.kde()
```

```python
plt.plot([1,2,3,4])
plt.show()
```

```python
plt.plot([1,3,2,7],[10,20,30,40])
plt.grid(True)
plt.show()
```

```python
plt.plot([10,20,30,40],[2,6,8,12],label='price')
plt.xlabel('X label')
plt.ylabel('Y label')
plt.legend()
plt.title('X Y graph')
plt.show()
```

```python
d=np.arange(0.,10.,0.4)
plt.plot(d,d*2,'r-',d,d*3,'b--')
plt.xlabel('X value')
plt.ylabel('Y value')
plt.title('X Y graph')
plt.show()
```

```python
d=np.arange(0.,10.,0.4)
plt.figure(1)

plt.subplot(211)
plt.plot(d,2*d,'r--')
plt.xlabel('X ')
plt.ylabel('Y ')
plt.title('Double graph')

plt.subplot(212)
plt.plot(d,d*-2,'b--')
plt.xlabel('X')
plt.ylabel('Y')

plt.show()
```

```python
plt.scatter(np.random.normal(5,3,1000),np.random.normal(3,5,1000))
plt.show()
```

```python
car = ('bmw','benz','kia','hd')
x_pos=np.arange(len(car))
speed= 80+60*np.random.randn(len(car))

fig,ax = plt.subplots()
ax.bar(x_pos,speed,align='center',color='grey')
ax.set_xticks(x_pos)
ax.set_xticklabels(car)
plt.show()
```
